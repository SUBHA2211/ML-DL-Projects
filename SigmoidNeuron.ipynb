Plotting Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

$S_{w,b}(x)=\frac{1}{1+e^{-(wx+b)}}$

```python
def sigmoid(w,x,b):
  return 1/(1+np.exp(-(w*x+b)))
```

```python
sigmoid(1,0.5,0)
```

```python
w = -1.2   # @param {type: "slider", min: -2, max: 2,step :0.1 }
b = -0.1    # @param {type: "slider", min: -2, max: 2,step :0.1 }
X = np.linspace(-10,10,100)
Y = sigmoid(w,X,b)
plt.plot(X,Y)
plt.show()
```

$S_{w_1,w_2}(x_1,x_2) = \frac{1}{1+e^{-(w_1x_1,w_2x_2+b)}}$

```python
def sigmoid_2d(x1,x2,w1,w2,b):
  return 1/(1+np.exp(-(w1*x1 + w2*x2 + b)))
```

```python
sigmoid_2d(1,0,0.5,0,0)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
X1 = np.linspace(-10,10,100)
X2 = np.linspace(-10,10,100)
XX1 , XX2 = np.meshgrid(X1,X2)
```

```python
w1 = 1.5
w2 = 0.5
b = 0
Y = sigmoid_2d(XX1,XX2,w1,w2,b)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.plot_surface(XX1,XX2,Y)
```

#Compute loss for given Dataset

```python
w_unknown = 0.5
b_unknown = 0.25
X = np.random.random(25)*20 - 10
Y = sigmoid(X,w_unknown,b_unknown)
```

```python
plt.plot(X,Y,'*')
plt.show()
```

```python
def compute_loss(X,Y,w_est,b_est):
  loss = 0
  for x,y in zip(X,Y):
    loss += (y-sigmoid(x,w_est,b_est))**2
  return loss
```

```python
W = np.linspace(-1,1,101)
B = np.linspace(-1,1,101)
WW,BB = np.meshgrid(W,B)

Loss = np.zeros(WW.shape)

```

```python
for i in range(WW.shape[0]):
  for j in range(WW.shape[1]):
    Loss[i,j] = compute_loss(X,Y,WW[i,j],BB[i,j])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.set_xlabel('W')
ax.set_ylabel('B')
ax.set_zlabel('LOSS')
ax.plot_surface(WW,BB,Loss)
```
